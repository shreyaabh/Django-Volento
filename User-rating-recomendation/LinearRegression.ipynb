```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
ratings=pd.read_csv("eventdataset.csv",index_col=0)
```

```python
pd.set_option('display.float_format', lambda x: '%.9f' % x)
```

```python
ratings.fillna(0, inplace=True)
ratings
```

```python
def standardize(row):
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row

df_std = ratings.apply(standardize).T
print(df_std)

sparse_df = sparse.csr_matrix(df_std.values)
corrMatrix = pd.DataFrame(cosine_similarity(sparse_df),index=ratings.columns,columns=ratings.columns)
corrMatrix
```

```python
corrMatrix = ratings.corr(method='pearson')
corrMatrix.head(6)
```

```python
def get_similar(event_name,rating):
    similar_score = corrMatrix[event_name]*(rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    return similar_score
```

```python
cleanup_lover = [("RiverCleanUp",5),("ClothDonation",1),("FoodDonation",1)]
similar_scores = pd.DataFrame()
for movie,rating in cleanup_lover:
    similar_scores = similar_scores.append(get_similar(movie,rating),ignore_index = True)

similar_scores.head(10)
```
